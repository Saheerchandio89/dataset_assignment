```python
plt.plot(df[['length_of_stay','flight_hour']],color='black', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='red', markersize=12)
plt.ylabel('length_of_stay')
plt.xlabel('flight_hour')
plt.title('Length of stay vs duration of flight')
plt.show()
```
